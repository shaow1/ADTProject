```python
print(df.columns)
df = df.dropna(subset=['Customer ID'])
df.to_csv('clean.csv', index=False)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.describe()
```

```python
df['TotalPrice'] = df['Quantity'] * df['Price']
```

```python
df.set_index('InvoiceDate').resample('M')['Invoice'].nunique().plot(title="Monthly Number of Invoices")
```

```python
df['Invoice_Month'] = df['InvoiceDate'].dt.month
df['Invoice_Month'].value_counts().sort_index().plot(kind='bar')
```

```python
df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False)
```
